In this notebook we will explore car dataset and we will be performorming Exploratory Data Analysis (EDA)

1.Descriptive Statistics:

What is the distribution of car prices in the dataset? Are there outliers? What is the average, median, and range of mileage? Which car make and model are the most and least common?

2.Trends Over Time:

How does the average price of cars vary by production year? What are the oldest and newest cars in the dataset?

3.Categorical Variables:

Which fuel type is the most popular, and does it vary by segment? What are the most common colors for cars on sale? How does the condition of cars correlate with mileage or price?

4.Segment Analysis:

What are the price ranges for different car segments? How accurate does the "segment" feature appear to be? Are there inconsistencies with the make/model?

5.Regression and Relationships Target Variable Analysis (Price):

What features have the strongest correlation with car price? Does mileage significantly affect the price of a car? Are certain brands priced higher on average than others?

6.Engine and Performance:

Does engine volume influence the price of a car? If so, is the relationship linear? Are diesel cars priced differently compared to petrol or electric cars?

7.Transmission and Drive Unit:

How does transmission type (manual vs. automatic) affect price? Are 4WD or AWD cars more expensive compared to front-wheel or rear-wheel drive cars?

8.Advanced Explorations Market Trends:

How does the price change for the same make and model as the production year increases? What is the depreciation rate for cars based on mileage?

9.Outliers and Anomalies:

Are there any unusual entries (e.g., very low or high prices)? Are some car makes/models priced unexpectedly high or low for their segment?

Feature Interactions:

10.How does fuel type interact with mileage and price? Is there a relationship between car color and price or segment?

Prediction:

Can a linear regression model predict the price of a car based on the other features? How does the model perform for different segments or conditions? Visualization Ideas Distribution plots for price and mileage. Box plots for price vs. categorical features (e.g., fuel type, color, transmission). Scatter plots to examine relationships (e.g., mileage vs. price, engine volume vs. price). Heatmaps to visualize correlations between numerical features.

```python
# DATA CLEANING
import pandas as pd

# Load the dataset correctly
df = pd.read_csv('cars.csv')

# Drop rows with missing values
df.dropna(inplace=True)

# Check for missing values again
print(df.isnull().sum())

# Print the cleaned DataFrame
print(df)
```

```python
# 1.What is the distribution of car prices in the dataset?

import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plot size
plt.figure(figsize=(10, 7))

# Plot the histogram
sns.histplot(df['priceUSD'], kde=False, bins=30, color='blue')

# Add title and labels
plt.title('Distribution of Car Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# 1.1 What is the average, median, and range of mileage?

# to calculate average of mileage I am goin to sum all values in mileage and divide the sum by number of values in this column

# Sum of mileage in the 'mileage(kilometers)' column
total_mileage = df['mileage(kilometers)'].sum()
print(total_mileage)

# 13745787871.092701 is our total

# Calculate number of rows in mileage(kilometers) column to get number of values
num_rows = df['mileage(kilometers)'].count()
print(num_rows)

# 56244 is our number of values

# 13745787871.092701 / 56244 = ?

print(13745787871.092701/56244)

print("average of mileage is 244395.63 kilometers")
```

```python
# 1.1 Calculate the median of the entire column
col_median = df['mileage(kilometers)'].median()
print(col_median)
```

```python
# 1.1 select a range of values from the mileage(kilometers) column
mileage_range = df.loc[(df['mileage(kilometers)'] >= 100000) & (df['mileage(kilometers)'] <= 200000), 'mileage(kilometers)']
print(mileage_range)
```

```python
import matplotlib.pyplot as plt

# 1.2 Count the number of cars for each make
car_make_counts = df['make'].value_counts()

# Get the top 5 most common car makes
top_10_makes = car_make_counts.head(10)

# Extract the makes (labels) and their counts
car_makes = top_10_makes.index
car_counts = top_10_makes.values

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(car_makes, car_counts, width=0.8, color=['red', 'green', 'yellow', 'indigo', 'violet' , 'saddlebrown', 'darkslategray', 'cyan', 'grey', 'black'])

# Label the axes and title
plt.xlabel('Car Makes')
plt.ylabel('Number of Cars')
plt.title('Top 10 Cars')

# Rotate x-axis labels for better visibility (optional)
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# 1.3 Which car makes are the least common?

import matplotlib.pyplot as plt

# Count the number of cars for each make
car_make_counts = df['make'].value_counts()

# Get the least 10 car makes
least_10_makes = car_make_counts.tail(10)

# Extract the makes (labels) and their counts
car_makes = least_10_makes.index
car_counts = least_10_makes.values

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(car_makes, car_counts, width=0.8, color=['red', 'green', 'yellow', 'indigo', 'violet' , 'saddlebrown', 'darkslategray', 'cyan', 'grey', 'black'])

# Label the axes and title
plt.xlabel('Car Makes')
plt.ylabel('Number of Cars')
plt.title('The least 10 Cars')

# Rotate x-axis labels for better visibility (optional)
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
#Trends Over Time:

# 2.How does the average price of cars vary by production year?

import matplotlib.pyplot as plt

# Calculate the average price of cars

average_price_by_year = df['priceUSD'].groupby(df["year"]).mean()
average_price_by_year = average_price_by_year.head()

# x axis values

x = average_price_by_year.index
y = average_price_by_year.values

# plotting the points
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)

# setting x and y axis range
# Adjust limits as needed based on data
plt.ylim(min(y), max(y) + 1000)  #Adding a margin ensures that the largest data point does not sit exactly on the edge
plt.xlim(min(x), max(x) + 1)

# naming the x axis
plt.xlabel('Year')
# naming the y axis
plt.ylabel('Price USD')
# giving a title to my graph
plt.title('Average Price of Cars By Production Year')

# function to show the plot
plt.show()
```

```python
# 2.1 What are the oldest cars in the dataset?

oldest_cars_by_year = df['make'].groupby(df['year']).min().head(20)

# Sort the years in ascending order
oldest_cars_by_year_sorted = oldest_cars_by_year.sort_index(ascending=True)

# Create a DataFrame from the sorted series
oldest_cars_df = pd.DataFrame({
    'year': oldest_cars_by_year_sorted.index,  # Years as the index
    'make': oldest_cars_by_year_sorted.values  # Car make for each year
})

# Print the DataFrame
print(oldest_cars_df)
```

```python
# 2.2 What are the newest cars in the dataset?

newest_cars_by_year = df['make'].groupby(df['year']).max().tail(20)  # Change head to tail to get the newest cars

# Sort the years in descending order
newest_cars_by_year_sorted = newest_cars_by_year.sort_index(ascending=False)

# Create a DataFrame from the sorted series
newest_cars_df = pd.DataFrame({
    'year': newest_cars_by_year_sorted.index,  # Years as the index
    'make': newest_cars_by_year_sorted.values  # Car make for each year
})

# Print the DataFrame
print(newest_cars_df)
```

```python
# 3.Categorical Variables:

#Which fuel type is the most popular, and does it vary by segment?

#To check if the most popular fuel type varies by car segment, we can group the data by the segment

# Group by 'segment' and count occurrences of each 'fuel_type'
fuel_type_by_segment = df.groupby('fuel_type')['segment'].value_counts()


# Creating a pivotable

pivot_table = pd.pivot_table(df,index = 'segment' , columns ='fuel_type', aggfunc='size', fill_value=0)
print(pivot_table)
```

```python
# 3.1 Categorical Variables:
#Which fuel type is the most popular, and does it vary by segment?

# Plot a grouped bar chart
pivot_table.plot(kind='bar', figsize=(10, 6), colormap='viridis')

# Add title and labels
plt.title("Fuel Type Popularity by Segment", fontsize=16)
plt.xlabel("Segment", fontsize=12)
plt.ylabel("Count of Cars", fontsize=12)
plt.legend(title='Fuel Type', fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

```python

# 3.3 Condition correlation with mileage and price
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='mileage(kilometers)', y='priceUSD', hue='condition', data=df)
plt.title('Mileage vs Price by Condition')
plt.show()
```

```python
# 4. Segment Analysis
# 4.1 Price ranges by segment
price_ranges_by_segment = df.groupby('segment')['priceUSD'].agg(['min', 'max', 'mean'])
print("Price Ranges by Segment:\n", price_ranges_by_segment)
```

```python

# 4. price ranges by segment
import matplotlib.pyplot as plt
import numpy as np

# Preparing data
segments = price_ranges_by_segment.index
mins = price_ranges_by_segment['min']
maxs = price_ranges_by_segment['max']
means = price_ranges_by_segment['mean']

# Bar width and positions
bar_width = 0.25
x = np.arange(len(segments))

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(x - bar_width, mins, width=bar_width, label='Min Price', color='red')
plt.bar(x, means, width=bar_width, label='Mean Price', color='grey')
plt.bar(x + bar_width, maxs, width=bar_width, label='Max Price', color='darkblue')

# Labels, title, and legend
plt.xlabel('Segment')
plt.ylabel('Price (USD)')
plt.title('Price Ranges by Segment')
plt.xticks(x, segments)
plt.legend()
plt.show()
```

```python
# 5.Regression and Relationships Target Variable Analysis (Price):

#What features have the strongest correlation with car price? Does mileage significantly affect the price of a car? Are certain brands priced higher on average than others?

# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Plot correlation heatmap
plt.figure(figsize=(10, 8))  # Adjust figure size
sns.heatmap(correlation_matrix, cmap="YlGnBu", annot=True, fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

INSIGHTS 1.Strongly Correlated Features with priceUSD:

year (correlation: +0.605): Newer cars generally cost more. volume(cm3) (correlation: +0.267): Larger engine volumes are weakly associated with higher prices.

2.Weak Negative Correlations:

mileage(kilometers) (correlation: -0.171): Higher mileage reduces price slightly, as expected for used cars

```python
## 5.Regression and Relationships Target Variable Analysis (Price):

#What features have the strongest correlation with car price?
#Does mileage significantly affect the price of a car?
#Are certain brands priced higher on average than others?

import matplotlib.pyplot as plt
import numpy as np

# Scatter plot of mileage vs. price with green points
x = df['mileage(kilometers)']
y = df['priceUSD']

plt.scatter(x, y, color='green', label='Green Points')

# Scatter plot of mileage vs. price with red points
# assuming x and y are the arrays that contain your data
x_data = np.array(df['mileage(kilometers)'])
y_data = np.array(df['priceUSD'])

# fitting a linear regression line
# Using degree 1 for a linear regression
coefficients = np.polyfit(x_data, y_data, 1)

# Unpack coefficients - coefficients[0] is slope, coefficients[1] is intercept
m, b = coefficients

# adding the regression line to the scatter plot
plt.plot(x_data, m * x_data + b, color='red', label='Regression Line') # Plot the regression line

plt.xlabel('Mileage (kilometers)') # Add labels and title for better visualization
plt.ylabel('Price (USD)')
plt.title('Mileage vs. Price with Regression Line')
plt.legend() # Display the legend
plt.show()
```

INSIGHTS: Mileage of car decreases the price

```python
5 #Are certain brands priced higher on average than others?

# Group by brand and calculate the average price
avg_price_by_brand = df.groupby('make')['priceUSD'].mean().sort_values(ascending=False)

# Plot the average price by brand
plt.figure(figsize=(12, 6))
avg_price_by_brand.plot(kind='bar', color='skyblue')

# Add title and labels
plt.title('Average Price by Brand')
plt.xlabel('Car Brand')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#6.Engine and Performance:

#Does engine volume influence the price of a car?
#If so, is the relationship linear?
#Are diesel cars priced differently compared to petrol or electric cars?


import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot with regression line
sns.regplot(x='volume(cm3)', y='priceUSD', data=df, scatter_kws={'color': 'green'}, line_kws={'color': 'red'})

# Add title and labels
plt.title("Engine Volume vs Price")
plt.xlabel("Engine Volume")
plt.ylabel("Price(USD)")
plt.grid()
```

```python
# Engine volume vs Price

#7.Transmission and Drive Unit:

#How does transmission type (manual vs. automatic) affect price? Are 4WD or AWD cars more expensive compared to front-wheel or rear-wheel drive cars?

# Group by transmission type and calculate price
avg_price_by_transmission = df.groupby('transmission')['priceUSD'].mean()

# Plot bar chart
plt.figure(figsize=(10, 6))
avg_price_by_transmission.plot(kind='bar', color=['red', 'green'])

# Add labels and title
plt.xlabel('Transmission Type')
plt.ylabel('Average Price (USD)')
plt.title('Average Price by Transmission Type')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Definitions

Mechanics (or Manual Transmission)  Manual Transmission (for a transmission system where the driver shifts gears manually).

Auto (or Automatic Transmission)  Automatic Transmission (for a transmission system that automatically shifts gears).

INSIGHTS: Average prices of automatic cars are higher compared to cars with manual gears system.

```python
#8.Advanced Explorations Market Trends:

#How does the price change for the same make and model as the production year increases? What is the depreciation rate for cars based on mileage?

# Depreciation based on mileage
depreciation = df.groupby('mileage(kilometers)')['priceUSD'].mean()
plt.plot(depreciation)

plt.xlabel('Mileage(Km)')
plt.ylabel('Average Price (USD)')
plt.title("Depreciation Rate Based on Mileage")
plt.show()
```

```python
#9.Outliers and Anomalies:

from scipy.stats import zscore

# Compute Z-scores for prices
df['z_score'] = zscore(df['priceUSD'])

# Define threshold (e.g., Z > 3 or Z < -3)
df['is_outlier'] = df['z_score'].abs() > 3

# Extract outliers
outliers = df[df['is_outlier']]

# Display the top 20 highest-priced outliers
top_20_outliers = outliers.nlargest(20, 'priceUSD')
print(top_20_outliers[['make', 'model', 'priceUSD']])

```

```python
# 10. Feature Interactions
#Fuel Type vs. Mileage and Price
#Does fuel type influence car mileage and price?

#Boxplot: Fuel type vs. Mileage
plt.figure(figsize=(12, 6))
sns.boxplot(x='fuel_type', y='mileage(kilometers)', data=df)
plt.title("Fuel Type vs. Mileage")
plt.xticks(rotation=45)
plt.show()

# Boxplot: Fuel type vs. Price
plt.figure(figsize=(12, 6))
sns.boxplot(x='fuel_type', y='priceUSD', data=df)
plt.title("Fuel Type vs. Price")
plt.xticks(rotation=45)
plt.show()
```

```python
# 10.1 Bar plot for average price by color
avg_price_by_color = df.groupby('color')['priceUSD'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_by_color.index, y=avg_price_by_color.values, palette="icefire")
plt.xticks(rotation=45)
plt.xlabel("Car Color")
plt.ylabel("Average Price (USD)")
plt.title("Average Car Price by Color")
plt.show()
```
